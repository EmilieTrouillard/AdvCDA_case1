```python
knn1 = neighbors.KNeighborsClassifier(2, weights='uniform')
knn1.fit(X_traindev1, y_traindev1)
knn2 = neighbors.KNeighborsClassifier(1, weights='uniform')
knn2.fit(X_traindev2, y_traindev2)
Z1 = knn1.predict(X_test)
Z2 = knn2.predict(X_test)
Z = [0] * len(Z1)
for i in range(len(Z1)):
    if Z1[i] == 0:
        Z[i] = 0
    else:
        Z[i] = Z2[i]
misclassification = 0
misclassified = [0,0,0,0,0]
total = [0,0,0,0,0]
for i, z in enumerate(Z):
    if z != y_test[i]:
        misclassification += 1
        misclassified[int(y_test[i])] += 1
    total[int(y_test[i])] += 1
misclassification = misclassification / len(Z)
accuracy = 1 - misclassification
accuracies = [1 - misclassified[i]/total[i] for i in range(5)]
print('total accuracy: ', accuracy)
print('class accuracies: ', accuracies)
```
